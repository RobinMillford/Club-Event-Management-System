```python
import pandas as pd
import pyodbc
import base64
from PIL import Image
```

```python
server = ''
database = 'Clubdb1'

cnxn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server}; \
                       SERVER=' + server +'; \
                       DATABASE=' + database +';\
                       Trusted_Connection=yes;')

cursor =cnxn.cursor()
```

```python
# Create events table
create_events_table = '''
CREATE TABLE Events (
    Event_Id VARCHAR(20) NOT NULL PRIMARY KEY,
    Club_Id VARCHAR(20) NOT NULL,
    Event_Name VARCHAR(100),
    Event_Picture NVARCHAR(MAX),
    Event_Type VARCHAR(20),
    Event_Day VARCHAR(10),
    Event_Date DATE,
    Event_Start_Time TIME,
    Event_End_Time TIME,
    Event_Venue VARCHAR(50),
    Event_Mail VARCHAR(100),  -- Adjust the length as needed
    Event_Phone_No VARCHAR(15),
    Event_Budget FLOAT,
    Event_Ticket VARCHAR(5),
    Event_url NVARCHAR(MAX),
    Event_Status VARCHAR(255),  -- Add Event_Status column
    Event_Details NVARCHAR(MAX)  -- Add Event_Details column
);'''

# Create clubs table
create_clubs_table = '''
CREATE TABLE Clubs (
    Club_Id VARCHAR(20) NOT NULL PRIMARY KEY,
    Club_Name VARCHAR(50) NOT NULL,
    Club_Type VARCHAR(15),
    Club_Picture NVARCHAR(MAX),
    Club_Faculty_Cordinator_Name VARCHAR(30),
    Club_Mail VARCHAR(20),
    Club_Activity VARCHAR(20),
    Club_Total_Events INT,
    Club_Details NVARCHAR(MAX),
    Club_Photos_Viwer NVARCHAR(MAX)
);'''

# Create user student table
create_user_student_table = '''
CREATE TABLE User_Student (
    User_Student_Name VARCHAR(30) NOT NULL,
    User_Student_Id VARCHAR(15) NOT NULL PRIMARY KEY,
    User_Student_Type VARCHAR(30),
    User_Student_Club_Id VARCHAR(20),
    FOREIGN KEY (User_Student_Club_Id) REFERENCES Clubs(Club_Id)
);'''

# Create user faculty table
create_user_faculty_table = '''
CREATE TABLE User_Faculty (
    User_Faculty_Name VARCHAR(30) NOT NULL,
    User_Faculty_Id VARCHAR(15) NOT NULL PRIMARY KEY,
    User_Faculty_Type VARCHAR(30),
    User_Faculty_Club_Id VARCHAR(20),
    FOREIGN KEY (User_Faculty_Club_Id) REFERENCES Clubs(Club_Id)
);'''

# Create user student welfare table
create_user_student_welfare_table = '''
CREATE TABLE User_Student_Welfare (
    User_Student_Welfare_Name VARCHAR(30) NOT NULL,
    User_Student_Welfare_Id VARCHAR(15) NOT NULL PRIMARY KEY,
    User_Student_Welfare_Type VARCHAR(30),
    User_Student_Welfare_Club_Id VARCHAR(20),
    User_Student_Welfare_Office VARCHAR(MAX)  -- Add User_Student_Welfare_Office column
);'''
```

```python
# Switch to the new database
cursor.execute("USE Clubdb1;")

# # Create the tables
cursor.execute(create_events_table)
cursor.execute(create_clubs_table)
cursor.execute(create_user_student_table)
cursor.execute(create_user_faculty_table)
cursor.execute(create_user_student_welfare_table)

# # Commit the changes
cnxn.commit()
```

```python
# Drop all tables in the database
drop_User_Student_Welfare = '''
IF OBJECT_ID('User_Student_Welfare', 'U') IS NOT NULL
    DROP TABLE User_Student_Welfare;
'''
cursor.execute(drop_User_Student_Welfare)
cnxn.commit()
# cursor.close()
# cnxn.close()
```
